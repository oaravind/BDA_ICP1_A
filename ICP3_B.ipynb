Why doesn't it work?

Hint: It's not numeric data.

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
car_sales = pd.read_csv('C:/Users/aravi/OneDrive/Desktop/Code_Help/car-sales.csv')
price_cleaned = car_sales['Price'].str.replace('[\$,]', '', regex=True)
price_no_decimals = price_cleaned.str.split('.').str[0]
car_sales['Price'] = price_no_decimals.astype(int)

car_sales['Total Sales'] = car_sales['Price'].cumsum()

car_sales['Date Sold'] = pd.date_range(start=pd.Timestamp.today(), periods=len(car_sales))

car_sales.head()
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column


# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)

# View the car_sales DataFrame

car_sales['Total Sales'] = car_sales['Price'].cumsum()

car_sales['Sale Date'] = pd.date_range(start=pd.Timestamp.today().normalize(), periods=len(car_sales))

car_sales.head()
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
car_sales.plot(x='Sale Date', y='Total Sales', kind='line', figsize=(10, 5))

plt.title("Cumulative Car Sales Over Time")
plt.xlabel("Sale Date")
plt.ylabel("Total Sales ($)")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Convert the 'Price' column to the integers


# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function

car_sales['Price'] = car_sales['Price'].astype(str)
car_sales['Price'] = car_sales['Price'].str.replace('[\$,]', '', regex=True)
car_sales['Price'] = car_sales['Price'].str.split('.').str[0].astype(int)

car_sales.plot(x='Odometer (KM)', y='Price', kind='scatter', figsize=(8, 5))

plt.title("Odometer Reading vs Car Price")
plt.xlabel("Odometer (KM)")
plt.ylabel("Price ($)")
plt.grid(True)

plt.show()
```

```python
# Create a NumPy array of random numbers of size (10, 4) and save it to X


# Turn the NumPy array X into a DataFrame with columns called ['a', 'b', 'c', 'd']


# Create a bar graph of the DataFrame
X = np.random.rand(10, 4)
df = pd.DataFrame(X, columns=['a', 'b', 'c', 'd'])
df.plot(kind='bar', figsize=(10, 6))

plt.title("Bar Plot of Random Data")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend(title="Columns")
plt.tight_layout()
plt.show()
```

```python
# Create a bar graph of the 'Make' and 'Odometer (KM)' columns in the car_sales DataFrame
make_odometer_avg = car_sales.groupby('Make')['Odometer (KM)'].mean()
make_odometer_avg.plot(kind='bar', figsize=(8, 5), color='skyblue')

plt.title("Average Odometer Reading by Car Make")
plt.xlabel("Car Make")
plt.ylabel("Average Odometer (KM)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# Create a histogram of the 'Odometer (KM)' column
car_sales['Odometer (KM)'].plot(kind='hist', bins=20, edgecolor='black', figsize=(8, 5))
plt.title("Histogram of Odometer Readings")
plt.xlabel("Odometer (KM)")
plt.ylabel("Number of Cars")
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# Replot the same figure as above except change the "cmap" parameter
# of scatter() to "winter"
# Also change the "color" parameter of axhline() to "red"

# Create the plot


# Plot the data


# Customize the plot


# Add a meanline
heart_disease = pd.read_csv("C:/Users/aravi/OneDrive/Desktop/Code_Help/heart-disease.csv")
plt.style.use('seaborn-whitegrid')

over_50 = heart_disease[heart_disease['age'] > 50]

fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(
    over_50['age'],
    over_50['chol'],
    c=over_50['target'],
    cmap='winter',  
    alpha=0.7,
    edgecolors='w',
    s=100
)

ax.set_title('Age vs Cholesterol for Patients Over 50 by Target')
ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol (mg/dl)')
ax.grid(True)

legend1 = ax.legend(*scatter.legend_elements(), title="Target")
ax.add_artist(legend1)

mean_chol = over_50['chol'].mean()
ax.axhline(mean_chol, color='red', linestyle='--', linewidth=2, label=f'Mean Cholesterol: {mean_chol:.1f}')  # changed here

ax.legend()
plt.show()
fig.savefig('C:/Users/aravi/OneDrive/Desktop/Code_Help/heart_disease_plot3.png', dpi=300, bbox_inches='tight')
```
